```python
from IPython.display import HTML

HTML('''
<!doctype html>
<html lang="en">
<head>
  <meta charset="utf-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1" />
  <title>Try Me — Fashion Store</title>
  <meta name="description" content="Try Me - Stylish online fashion store for Men, Women, Kids & Accessories" />
  <style>
    /* ====== RESET & BASE ====== */
    :root{
      --bg:#0f1724;
      --card:#0f1724;
      --glass: rgba(255,255,255,0.06);
      --accent1: #ff7eb3; /* pink */
      --accent2: #7afcff; /* aqua */
      --accent3: #ffd36e; /* gold */
      --text: #eef2ff;
      --muted: rgba(238,242,255,0.7);
      --success: #4ade80;
      --danger: #ff6b6b;
      font-family: Inter, system-ui, -apple-system, 'Segoe UI', Roboto, 'Helvetica Neue', Arial;
    }
    *{
      box-sizing:border-box;
      margin:0;
      padding:0
    }
    html,body{
      height:100%;
      background:linear-gradient(135deg,#061225 0%, #081a2b 40%, #071018 100%);
      color:var(--text);
    }
    a{
      text-decoration:none;
      color:inherit
    }
    .container{
      max-width:1200px;
      margin:30px auto;
      padding:20px
    }

    /* ====== NAV ====== */
    header{
      display:flex;
      align-items:center;
      justify-content:space-between;
      padding:14px;
      border-radius:14px;
      background:linear-gradient(90deg,rgba(255,255,255,0.02), rgba(255,255,255,0.01));
      backdrop-filter:blur(6px);
      box-shadow:0 6px 20px rgba(2,6,23,0.6)
    }
    .brand{
      display:flex;
      gap:12px;
      align-items:center
    }
    .logo{
      width:54px;
      height:54px;
      border-radius:12px;
      background:linear-gradient(135deg,var(--accent1),var(--accent2));
      display:flex;
      align-items:center;
      justify-content:center;
      font-weight:800;
      font-size:18px;
      color:#08121a;
      box-shadow:0 6px 18px rgba(124,58,237,0.15)
    }
    nav{
      display:flex;
      gap:18px;
      align-items:center
    }
    nav a{
      padding:8px 12px;
      border-radius:10px;
      color:var(--muted);
      font-weight:600
    }
    nav a:hover{
      background:var(--glass);
      color:var(--text)
    }
    .actions{
      display:flex;
      gap:12px;
      align-items:center
    }
    .btn{
      padding:10px 14px;
      border-radius:10px;
      font-weight:700;
      cursor:pointer;
      border:0
    }
    .btn-primary{
      background:linear-gradient(90deg,var(--accent1),var(--accent2));
      color:#08121a
    }
    .btn-ghost{
      background:transparent;
      border:1px solid rgba(255,255,255,0.06);
      color:var(--muted)
    }
    .cart-bubble{
      background:var(--accent3);
      color:#08121a;
      padding:6px 8px;
      border-radius:999px;
      font-weight:800
    }

    /* ====== LAYOUT ====== */
    .grid{
      display:grid;
      grid-template-columns:1fr 320px;
      gap:20px;
      margin-top:22px
    }

    /* ====== HERO ====== */
    .hero{
      padding:28px;
      border-radius:14px;
      background:linear-gradient(135deg, rgba(255,126,179,0.06), rgba(122,252,255,0.03));
      display:flex;
      gap:18px;
      align-items:center
    }
    .hero-left{
      flex:1
    }
    .hero h1{
      font-size:28px;
      margin-bottom:6px;
      letter-spacing:0.2px
    }
    .hero p{
      color:var(--muted);
      margin-bottom:12px
    }
    .search{
      display:flex;
      gap:10px;
      align-items:center
    }
    .search input{
      flex:1;
      padding:12px;
      border-radius:12px;
      border:0;
      background:rgba(255,255,255,0.03);
      color:var(--text)
    }

    /* ====== PRODUCT LIST ====== */
    .products{
      display:grid;
      grid-template-columns:repeat(auto-fill,minmax(220px,1fr));
      gap:20px;
      margin-top:20px
    }
    .card{
      border-radius:14px;
      background:linear-gradient(180deg, rgba(255,255,255,0.02), rgba(255,255,255,0.01));
      padding:14px;
      box-shadow:0 8px 30px rgba(2,6,23,0.5);
      display:flex;
      flex-direction:column;
      gap:10px
    }
    .img-wrap{
      height:180px;
      border-radius:12px;
      background:linear-gradient(135deg,var(--accent2),var(--accent1));
      display:flex;
      align-items:center;
      justify-content:center;
      color:#05101a;
      font-weight:700;
      font-size:20px
    }
    .price-row{
      display:flex;
      justify-content:space-between;
      align-items:center
    }
    .price{
      font-weight:900
    }
    .muted{
      color:var(--muted);
      font-size:13px
    }

    /* ====== SIDEBAR CART ====== */
    .sidebar{
      position:sticky;
      top:30px;
      padding:18px;
      border-radius:12px;
      background:linear-gradient(180deg, rgba(255,255,255,0.02), rgba(255,255,255,0.01));
      box-shadow:0 8px 30px rgba(2,6,23,0.5)
    }
    .cart-item{
      display:flex;
      gap:12px;
      align-items:center;
      padding:8px 0;
      border-bottom:1px dashed rgba(255,255,255,0.03)
    }
    .qty{
      background:rgba(0,0,0,0.3);
      padding:6px 10px;
      border-radius:8px
    }
    .total-row{
      display:flex;
      justify-content:space-between;
      align-items:center;
      margin-top:12px;
      font-weight:800
    }

    /* ====== LOGIN & PAYMENT MODAL (SIMPLE) ====== */
    .panel{
      max-width:420px;
      margin:30px auto;
      padding:18px;
      border-radius:12px;
      background:linear-gradient(180deg, rgba(255,255,255,0.02), rgba(255,255,255,0.01));
      box-shadow:0 10px 40px rgba(2,6,23,0.6)
    }
    form{
      display:flex;
      flex-direction:column;
      gap:12px
    }
    input[type=text],input[type=email],input[type=password],input[type=number]{
      padding:12px;
      border-radius:10px;
      border:0;
      background:rgba(255,255,255,0.03);
      color:var(--text)
    }
    label{
      color:var(--muted);
      font-size:13px
    }
    .small{
      font-size:13px;
      color:var(--muted)
    }

    /* ====== FOOTER ====== */
    footer{
      margin-top:26px;
      padding:20px;
      border-radius:12px;
      text-align:center;
      color:var(--muted);
      background:transparent
    }

    /* ====== RESPONSIVE ====== */
    @media(max-width:900px){
      .grid{grid-template-columns:1fr}
      .logo{width:46px;height:46px}
      header{flex-direction:column;gap:12px}
    }
  </style>
</head>
<body>
  <div class="container">
    <header>
      <div class="brand">
        <div class="logo">TryMe</div>
        <div>
          <div style="font-weight:800">Try Me</div>
          <div class="small">Style your life</div>
        </div>
      </div>

      <nav>
        <a href="#" onclick="showView('home')">Home</a>
        <a href="#" onclick="showView('women')">Women</a>
        <a href="#" onclick="showView('men')">Men</a>
        <a href="#" onclick="showView('kids')">Kids</a>
      </nav>

      <div class="actions">
        <button class="btn btn-ghost" onclick="showLogin()">Login</button>
        <button class="btn btn-primary" onclick="showView('cart')">Cart <span id="cartCount" class="cart-bubble">0</span></button>
      </div>
    </header>

    <!-- VIEWS -->
    <main>
      <section id="homeView">
        <div class="hero">
          <div class="hero-left">
            <h1>New season. Fresh looks.</h1>
            <p>Discover colorful outfits for Men, Women & Kids. Free shipping on orders above ₹1999.</p>
            <div class="search">
              <input id="searchInput" placeholder="Search for dresses, shirts, accessories..."/>
              <button class="btn btn-primary" onclick="performSearch()">Search</button>
            </div>
          </div>
          <div style="width:320px; text-align:right">
            <img src="" alt="" style="width:260px;opacity:0.95;border-radius:12px;filter:drop-shadow(0 10px 30px rgba(0,0,0,0.6))" />
          </div>
        </div>

        <div class="grid">
          <div>
            <h3 style="margin-top:14px">Popular picks</h3>
            <div id="productList" class="products"></div>
          </div>

          <aside class="sidebar">
            <h3>Your Cart</h3>
            <div id="cartList"></div>
            <div class="total-row">
              <div>Total</div>
              <div id="cartTotal">₹0</div>
            </div>
            <div style="margin-top:12px;display:flex;gap:10px">
              <button class="btn btn-ghost" onclick="clearCart()">Clear</button>
              <button class="btn btn-primary" onclick="checkout()">Checkout</button>
            </div>
          </aside>
        </div>
      </section>

      <!-- LOGIN PANEL -->
      <section id="loginPanel" style="display:none">
        <div class="panel">
          <h2>Welcome back</h2>
          <p class="small">Login to access your orders and preferences</p>
          <form id="loginForm" onsubmit="return login(event)">
            <label>Email</label>
            <input type="email" id="email" required placeholder="you@example.com" />
            <label>Password</label>
            <input type="password" id="password" required placeholder="••••••••" />
            <div style="display:flex;gap:8px;justify-content:flex-end">
              <button type="button" class="btn btn-ghost" onclick="showView('home')">Cancel</button>
              <button type="submit" class="btn btn-primary">Login</button>
            </div>
          </form>
        </div>
      </section>

      <!-- CART / CHECKOUT VIEW -->
      <section id="checkoutPanel" style="display:none">
        <div class="panel">
          <h2>Checkout</h2>
          <div id="checkoutItems"></div>
          <div class="total-row" style="margin-top:10px"><div>Amount</div><div id="checkoutAmount">₹0</div></div>
          <hr style="border:none;border-top:1px dashed rgba(255,255,255,0.03);margin:12px 0" />
          <h3>Payment</h3>
          <form id="paymentForm" onsubmit="return pay(event)">
            <label>Name on card</label>
            <input type="text" id="cardName" required placeholder="Full name" />
            <label>Card number</label>
            <input type="text" id="cardNumber" maxlength="19" required placeholder="1234 5678 9012 3456" />
            <div style="display:flex;gap:8px">
              <div style="flex:1">
                <label>Expiry (MM/YY)</label>
                <input type="text" id="expiry" maxlength="5" required placeholder="08/26" />
              </div>
              <div style="width:120px">
                <label>CVV</label>
                <input type="password" id="cvv" maxlength="4" required placeholder="123" />
              </div>
            </div>
            <div style="display:flex;gap:8px;justify-content:flex-end;margin-top:12px">
              <button type="button" class="btn btn-ghost" onclick="showView('home')">Back</button>
              <button type="submit" class="btn btn-primary">Pay now</button>
            </div>
          </form>
        </div>
      </section>

      <footer>
        © <span id="year"></span> Try Me. Crafted with ❤️
      </footer>
    </main>
  </div>

  <script>
    // ======= Sample product data =======
    const products = [
      {id:1,title:'Floral Summer Dress',price:1299,cat:'women'},
      {id:2,title:'Men Classic Shirt',price:899,cat:'men'},
      {id:3,title:'Kids Cartoon Tee',price:499,cat:'kids'},
      {id:4,title:'Women Silk Sari',price:3499,cat:'women'},
      {id:5,title:'Men Denim Jacket',price:2199,cat:'men'},
      {id:6,title:'Golden Earrings',price:699,cat:'accessories'}
    ];

    let cart = {};

    // ======= UI helpers =======
    function $(id){return document.getElementById(id)}
    function showView(name){
      // hide all
      ['homeView','loginPanel','checkoutPanel'].forEach(id=>document.getElementById(id).style.display='none');
      if(name==='home') document.getElementById('homeView').style.display='block';
      else if(name==='login') document.getElementById('loginPanel').style.display='block';
      else if(name==='cart') { document.getElementById('homeView').style.display='block'; scrollToCart(); }
      else if(name==='checkout') { document.getElementById('checkoutPanel').style.display='block'; }
    }

    function showLogin(){showView('login')}
    function scrollToCart(){window.scrollTo({top:800,behavior:'smooth'})}

    // ======= Render products & cart =======
    function renderProducts(list=products){
      const container = $('productList'); container.innerHTML='';
      list.forEach(p=>{
        const el = document.createElement('div'); el.className='card';
        el.innerHTML = `
          <div class="img-wrap">${p.title.split(' ')[0]}</div>
          <div style="flex:1">
            <div style="font-weight:800">${p.title}</div>
            <div class="muted">Category: ${p.cat}</div>
          </div>
          <div class="price-row">
            <div class="price">₹${p.price}</div>
            <div>
              <button class="btn btn-ghost" onclick="addToCart(${p.id})">Add</button>
            </div>
          </div>
        `;
        container.appendChild(el);
      });
    }

    function renderCart(){
      const list = $('cartList'); list.innerHTML='';
      let total=0; let count=0;
      Object.values(cart).forEach(item=>{
        total += item.price * item.qty; count += item.qty;
        const div = document.createElement('div'); div.className='cart-item';
        div.innerHTML = `
          <div style="width:60px;height:60px;border-radius:8px;background:linear-gradient(135deg,var(--accent2),var(--accent1));display:flex;align-items:center;justify-content:center;font-weight:800">${item.title.split(' ')[0]}</div>
          <div style="flex:1">
            <div style="font-weight:700">${item.title}</div>
            <div class="muted">₹${item.price} x ${item.qty}</div>
          </div>
          <div style="text-align:right">
            <div class="qty">${item.qty}</div>
            <div style="margin-top:6px"><button class="btn btn-ghost" onclick="removeFromCart(${item.id})">-</button></div>
          </div>
        `;
        list.appendChild(div);
      });
      $('cartTotal').innerText = '₹' + total;
      $('cartCount').innerText = count;
    }

    function addToCart(id){
      const p = products.find(x=>x.id===id); if(!p) return;
      if(cart[id]) cart[id].qty++;
      else cart[id] = {...p,qty:1};
      renderCart();
    }
    function removeFromCart(id){
      if(!cart[id]) return; cart[id].qty--; if(cart[id].qty<=0) delete cart[id]; renderCart();
    }
    function clearCart(){cart={};renderCart();}

    // ======= Simple search =======
    function performSearch(){
      const q = $('searchInput').value.trim().toLowerCase();
      if(!q) { renderProducts(); return }
      const filtered = products.filter(p=>p.title.toLowerCase().includes(q) || p.cat.toLowerCase().includes(q));
      renderProducts(filtered);
    }

    // ======= Login (mock) =======
    function login(e){
      e.preventDefault();
      const email = $('email').value.trim();
      // mock success
      alert('Logged in as ' + email);
      showView('home');
    }

    // ======= Checkout & Payment (mock) =======
    function checkout(){
      // show checkout panel and populate items
      if(Object.keys(cart).length===0){ alert('Your cart is empty'); return }
      const el = $('checkoutItems'); el.innerHTML=''; let amount=0;
      Object.values(cart).forEach(it=>{ amount += it.price*it.qty; const div = document.createElement('div'); div.style.marginBottom='8px'; div.innerHTML = `${it.title} — ${it.qty} x ₹${it.price}`; el.appendChild(div); });
      $('checkoutAmount').innerText = '₹' + amount; showView('checkout');
    }

    function pay(e){
      e.preventDefault();
      // basic validation
      const name = $('cardName').value.trim();
      const num = $('cardNumber').value.replace(/\s/g,'');
      if(num.length<12){ alert('Enter a valid card number'); return }
      // mock processing
      alert('Payment successful! Thank you ' + name);
      clearCart(); showView('home');
    }

    // ======= utilities =======
    function showLogin(){ $('loginPanel').style.display='block'; $('homeView').style.display='none'; }

    // initial render
    renderProducts(); renderCart(); $('year').innerText = new Date().getFullYear();

    // small helpers to format card input
    document.addEventListener('input',function(e){
      if(e.target.id==='cardNumber'){
        e.target.value = e.target.value.replace(/[^0-9]/g,'').replace(/(.{4})/g,'$1 ').trim();
      }
      if(e.target.id==='expiry'){
        e.target.value = e.target.value.replace(/[^0-9]/g,'').replace(/(.{2})/g,'$1/').slice(0,5);
      }
    });

  </script>
</body>
</html>
''')
```
